# 03 - Testando modelo via API FastAPI

Este notebook testa o endpoint `/predict` da API `fastapi-clima`, 
envia uma requisição com dados climáticos e verifica:

- se a API responde corretamente
- se o modelo retorna uma previsão numérica
- se a previsão é enviada ao ThingsBoard (device de previsão)


```python
import requests
import json
from pprint import pprint
```

```python
API_URL = "http://host.docker.internal:8000/predict"
API_URL
```

```python
payload_exemplo = {
    "temp_ar": 28.5,
    "radiacao": 900,
    "vento_vel": 2.3,
    "precipitacao": 0.0,
    "pressao": 1010.2,
    "cluster": 0   # só um valor de teste, por enquanto
}

payload_exemplo

```

```python
def chamar_api_predict(payload):
    """
    Envia um POST para o endpoint /predict da API fastapi-clima
    e retorna o JSON de resposta.
    """
    try:
        r = requests.post(API_URL, json=payload, timeout=10)
        print(f"Status code: {r.status_code}")
        try:
            resposta_json = r.json()
        except json.JSONDecodeError:
            print("Resposta não é JSON. Conteúdo bruto:")
            print(r.text)
            return None
        
        pprint(resposta_json)
        return resposta_json
    except Exception as e:
        print("Erro ao chamar a API:", type(e), e)
        return None
```

```python
resposta = chamar_api_predict(payload_exemplo)
```
